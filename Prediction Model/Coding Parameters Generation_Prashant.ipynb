```python
df_coding = df.iloc[:,[8,16,31,35]]
df_coding.head()
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
# Implementing TSNE for visualization of coding
df_coding_tsne = TSNE(n_components=1).fit_transform(df_coding)
# Implementing PCA to reduce dimensions
pca = PCA(n_components=1)
df_coding_pca = pca.fit_transform(df_coding)
```
