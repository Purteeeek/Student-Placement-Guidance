```python
path = "../Cleaned-Data/16,17,18.xlsx"
```

```python
df = pd.read_excel(path)
```

```python
df_logical = df.iloc[:,[3,16,26,31]]
df_logical.head()
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
# Implementing TSNE for visualization of coding
df_logical_tsne = TSNE(n_components=1).fit_transform(df_logical)
# Implementing PCA to reduce dimensions
pca = PCA(n_components=1)
df_logical_pca = pca.fit_transform(df_logical)
```
