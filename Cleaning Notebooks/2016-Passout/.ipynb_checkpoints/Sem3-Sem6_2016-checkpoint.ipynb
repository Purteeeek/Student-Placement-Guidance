```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
% matplotlib inline
```

```python
sem3_2016 = '../../Cleaned-Data/2016-Passouts/Sem-3_2016_Passout_cleaned.xlsx'
sem4_2016 = '../../Cleaned-Data/2016-Passouts/Sem-4_2016_Passout_cleaned.xlsx'
sem5_2016 = '../../Cleaned-Data/2016-Passouts/Sem-5_2016_Passout_cleaned.xlsx'
sem6_2016 = '../../Cleaned-Data/2016-Passouts/Sem-6_2016_Passout_cleaned.xlsx'
```

```python
sem3_df = pd.read_excel(sem3_2016)
sem4_df = pd.read_excel(sem4_2016)
sem5_df = pd.read_excel(sem5_2016)
sem6_df = pd.read_excel(sem6_2016)
```

```python
sem3_df.info()
```

```python
sem4_df.info()
```

```python
sem5_df.info()
```

```python
sem6_df.info()
```

```python
merge = pd.concat([sem3_df.set_index('Name'), sem4_df.set_index('Name'), sem5_df.set_index('Name'), sem6_df.set_index('Name')], axis=1, join='inner')
merge.head()
```

```python
merge.info()
```

```python
merge.describe()
```

```python
merge = merge.drop(['Sno'], axis=1)
merge.head()
```

```python
writer = pd.ExcelWriter('../../Cleaned-Data/2016-Passouts/2016_merged.xlsx')
```

```python
merge.to_excel(writer,"2016_merged")
writer.save()
```

```python
merged_path = '../../Cleaned-Data/2016-Passouts/2016_merged.xlsx'
placed_path = '../../Cleaned-Data/2016-Passouts/2016.xlsx'
```

```python
merged = pd.read_excel(merged_path)
placed = pd.read_excel(placed_path)
```

```python
merged.head(3)
```

```python
placed.head(3)
```

```python
merged_placed = pd.merge(merged, placed, on="Name", how="left")
```

```python
merged_placed.head(3)
```

```python
writer = pd.ExcelWriter('../../Cleaned-Data/2016-Passouts/2016_merged.xlsx')
```
