```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
% matplotlib inline
```

```python
sem3_2016 = '../../Cleaned-Data/2016-Passouts/Sem-3_2016_Passout_cleaned.xlsx'
sem4_2016 = '../../Cleaned-Data/2016-Passouts/Sem-4_2016_Passout_cleaned.xlsx'
sem5_2016 = '../../Cleaned-Data/2016-Passouts/Sem-5_2016_Passout_cleaned.xlsx'
sem6_2016 = '../../Cleaned-Data/2016-Passouts/Sem-6_2016_Passout_cleaned.xlsx'
```

```python
sem3_df = pd.read_excel(sem3_2016)
sem4_df = pd.read_excel(sem4_2016)
sem5_df = pd.read_excel(sem5_2016)
sem6_df = pd.read_excel(sem6_2016)
```

```python
merge = pd.concat([sem3_df.set_index('Name'), sem4_df.set_index('Name'), sem5_df.set_index('Name'), sem6_df.set_index('Name')], axis=1, join='inner')
merge.head()
```

```python
merge = merge.drop(['SNO', 'Sno'], axis=1)
merge.head()
```

```python
path2 = "/Users/prashant/Downloads/Student-Placement-Guidance/Cleaned-Data/2016-Passouts/2016.xlsx"
placement = pd.read_excel(path2)
placement.head()
```

```python
merge_with_company  = pd.merge(merge, placement, on = 'Name', how='left')
merge_with_company
```

```python
writer = pd.ExcelWriter('2016_merged.xlsx')
```

```python
merge_with_company.to_excel(writer,"2016_merged")
writer.save()
```
