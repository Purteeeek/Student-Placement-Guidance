```python
sem3_df = pd.read_excel(sem3)
sem4_df = pd.read_excel(sem4)
sem5_df = pd.read_excel(sem5)
sem6_df = pd.read_excel(sem6)
```

```python
#sem3_df.info()
```

```python
sem3_df = sem3_df.drop(['EXAM SEAT NO..1', 'AMIII_Grade', 'AMIII_TW_Grade', 'EDLC_Grade.1', 'EDLC_Grade', 'EDLC_Practical_Grade','DSGT_Grade', 'DSGT_TW_Grade'], axis=1)
sem3_df = sem3_df.drop(['DLDA_Grade', 'DLDA_TW_Grade', 'DSF_Grade', 'DSF_Grade.1', 'DSF_Practical_Grade', 'COA_Grade', 'COA_TW_Grade', 'PCT_Grade', 'Marks_in_words'], axis=1)
sem3_df = sem3_df.rename(columns = {'TOTAL': 'Total III', 'REMARKS':'REMARKS 3'})
sem3_df.info()
```

```python
sem4_df.info()
```

```python
sem4_df = sem4_df.drop(['EXAM SEAT NO..1', 'AMIV_Grade', 'AMIV_TW_Grade', 'ADC_Grade', 'ADC_TW_Grade', 'DBMS_Grade', 'DBMS_TW_Grade', 'DBMS_Practical_Grade'], axis = 1)
sem4_df = sem4_df.drop(['CG_Grade', 'CG_TW_Grade', 'CG_Practical_Grade', 'AAD_Grade', 'AAD_TW_Grade', 'AAD_Practical_Grade', 'OS_Grade', 'OS_TW_Grade', 'OS_Practical_Grade', 'Unnamed: 39'], axis = 1)
sem4_df = sem4_df.rename(columns = {'GNT' : 'GNT 3-4', 'REMARK':'REMARK 4'})
sem4_df.info()
```

```python
sem5_df.info()
```

```python
sem5_df = sem5_df.drop(['CN_Grade', 'CN_TW_Grade', 'CN_Practical_Grade', 'ADS_Grade', 'ADS_TW_Grade', 'ADS_Practical_Grade', 'MP_Grade', 'MP_TW_Grade', 'MP_Practical_Grade'], axis=1)
sem5_df = sem5_df.drop(['TCS_Grade', 'TCS_TW_Grade', 'WE_Grade', 'WE_Practical_Grade', 'EVS_Grade', 'EVS_TW_Grade', 'Total_in_words', 'Unnamed: 37'], axis = 1)
sem5_df = sem5_df.rename(columns = {'TOTAL': 'Total V', 'REMARK':'REMARK 5', 'NAME OF THE CANDIDATE' :'NAME OF CANDIDATE'})
#sem5_df.head(1)
sem5_df.info()
```

```python
sem6_df.info()
```

```python
sem6_df = sem6_df.drop(['EXAM SEAT NO..1', "ACN_Grade", "ACN_TW_Grade", "ACN_P&O_Grade", "SPCC_Grade", "SPCC_TW_Grade", "SPCC_P&O_Grade", "OOSE_Grade","OOSE_TW_Grade", "OOSE_P&O_Grade"], axis=1)
sem6_df = sem6_df.drop(["AMP_Grade", "AMP_TW_Grade","AMP_P&O_Grade", "DWM_Grade", "DWM_TW_Grade", "DWM_P&O_Grade", "SEMR_Grade", "SEMR_P&O_Grade", "Unnamed: 41"], axis=1)
sem6_df = sem6_df.rename(columns = {'GNT' : 'GNT 5-6', 'REMARK': 'REMARK 6', 'NAME OF THE CANDIDATE': 'NAME OF CANDIDATE'})
#sem6_df.head(1)
sem6_df.info()
```

```python
sem3_to_sem6_merge_df = pd.concat([sem3_df.set_index('NAME OF CANDIDATE'), sem4_df.set_index('NAME OF CANDIDATE'), sem5_df.set_index('NAME OF CANDIDATE'), sem6_df.set_index('NAME OF CANDIDATE')], axis=1, join='inner')
sem3_to_sem6_merge_df.head(10)
```
