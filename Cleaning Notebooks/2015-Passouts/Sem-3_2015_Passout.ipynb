```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
% matplotlib inline
```

```python
path = "../../Data/Exam Data/2015/CMPN SEM III DEC12.xlsx"
```

```python
sem_3_2015_df = pd.read_excel(path)
```

```python
sem_3_2015_df.info()
```

```python
#removed the rows where seatnumber was not present
sem_3_2015_df.dropna(subset=['EXAM SEAT NO.'],inplace=True)
```

```python
#removed the rows where all the values were NaN
sem_3_2015_df.dropna(axis=0, how='all',inplace=True)
```

```python
#made seatnumber as the index of the dataframe as there was no index.
sem_3_2015_df.index = sem_3_2015_df['EXAM SEAT NO.']
sem_3_2015_df.head(1)
```

```python
sem_3_2015_df.info()
```

```python
sem_3_2015_df.columns.values[[3,4,5]] = ["AMIII_Grade","AMIII_TW","AMIII_TW_Grade"]
sem_3_2015_df.columns.values[[7,8,9,10,11]] = ["EDLC_Grade","EDLC_TW","EDLC_Grade","EDLC_Practical","EDLC_Practical_Grade"]
sem_3_2015_df.columns.values[[13,14,15]] = ["DSGT_Grade","DSGT_TW","DSGT_TW_Grade"]
sem_3_2015_df.columns.values[[17,18,19]] = ["DLDA_Grade","DLDA_TW","DLDA_TW_Grade"]
sem_3_2015_df.columns.values[[21,22,23,24,25]] = ["DSF_Grade","DSF_TW","DSF_Grade","DSF_Practical","DSF_Practical_Grade"]
sem_3_2015_df.columns.values[[27,28,29]] = ["COA_Grade","COA_TW","COA_TW_Grade"]
sem_3_2015_df.columns.values[[31,34,35]] = ["PCT_Grade","Marks_in_words","Marks_numerical"]
```

```python
sem_3_2015_df = sem_3_2015_df.rename_axis(lambda x: x, 1, True)
sem_3_2015_df.head(1)
```

```python
sem_3_2015_df.info()
```

```python
sem_3_2015_df['AM III'] = pd.to_numeric(sem_3_2015_df["AM III"], errors='coerce')
sem_3_2015_df['AMIII_TW'] = pd.to_numeric(sem_3_2015_df["AMIII_TW"], errors='coerce')
sem_3_2015_df['EDLC'] = pd.to_numeric(sem_3_2015_df["EDLC"], errors='coerce')
sem_3_2015_df['EDLC_TW'] = pd.to_numeric(sem_3_2015_df["EDLC_TW"], errors='coerce')
sem_3_2015_df['EDLC_Practical'] = pd.to_numeric(sem_3_2015_df["EDLC_Practical"], errors='coerce')
sem_3_2015_df['DSGT'] = pd.to_numeric(sem_3_2015_df["DSGT"], errors='coerce')
sem_3_2015_df['DSGT_TW'] = pd.to_numeric(sem_3_2015_df["DSGT_TW"], errors='coerce')
sem_3_2015_df['DLDA'] = pd.to_numeric(sem_3_2015_df["DLDA"], errors='coerce')
sem_3_2015_df['DLDA'] = pd.to_numeric(sem_3_2015_df["DLDA_TW"], errors='coerce')
sem_3_2015_df['DSF'] = pd.to_numeric(sem_3_2015_df["DSF"], errors='coerce')
sem_3_2015_df['DSF_TW'] = pd.to_numeric(sem_3_2015_df["DSF_TW"], errors='coerce')
sem_3_2015_df['DSF_Practical'] = pd.to_numeric(sem_3_2015_df["DSF_Practical"], errors='coerce')
sem_3_2015_df['COA'] = pd.to_numeric(sem_3_2015_df["COA"], errors='coerce')
sem_3_2015_df['COA_TW'] = pd.to_numeric(sem_3_2015_df["COA_TW"], errors='coerce')
sem_3_2015_df['PCT'] = pd.to_numeric(sem_3_2015_df["PCT"], errors='coerce')
sem_3_2015_df['TOTAL'] = pd.to_numeric(sem_3_2015_df["TOTAL"], errors='coerce')
sem_3_2015_df['Marks_numerical'] = pd.to_numeric(sem_3_2015_df["Marks_numerical"], errors='coerce')
```
