```python
sem_6_2015_df.dropna(subset=['EXAM SEAT NO.'],inplace=True)
```

```python
sem_6_2015_df.index = sem_6_2015_df['EXAM SEAT NO.']
```

```python
sem_6_2015_df.info()
```

```python
sem_6_2015_df.columns.values[[3,4,5,6,7]] = ['ACN_Grade','ACN_TW','ACN_TW_Grade','ACN_P&O','ACN_P&O_Grade']
sem_6_2015_df.columns.values[[9,10,11,12,13]] = ['SPCC_Grade','SPCC_TW','SPCC_TW_Grade','SPCC_P&O','SPCC_P&O_Grade']
sem_6_2015_df.columns.values[[15,16,17,18,19]] = ['OOSE_Grade','OOSE_TW','OOSE_TW_Grade','OOSE_P&O','OOSE_P&O_Grade']
sem_6_2015_df.columns.values[[21,22,23,24,25]] = ['AMP_Grade','AMP_TW','AMP_TW_Grade','AMP_P&O','AMP_P&O_Grade']
sem_6_2015_df.columns.values[[27,28,29,30,31]] = ['DWM_Grade','DWM_TW','DWM_TW_Grade','DWM_P&O','DWM_P&O_Grade']
sem_6_2015_df.columns.values[[33,34,35]] = ['SEMR_Grade','SEMR_P&O','SEMR_P&O_Grade']
```

```python
sem_6_2015_df
```

```python
sem_6_2015_df['ACN'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['ACN_TW'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['ACN_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['SPCC'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['SPCC_TW'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['SPCC_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['OOSE'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['OOSE_TW'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['OOSE_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['AMP'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['AMP_TW'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['AMP_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['DWM'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['DWM_TW'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['DWM_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['SEMR'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['SEMR_P&O'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

sem_6_2015_df['TOT VI'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['TOT V'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
sem_6_2015_df['GNT'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

#removes non digit characters from pandas column AMIV
sem_6_2015_df
```

```python
sem_6_2015_df['ACN'] = pd.to_numeric(sem_6_2015_df["ACN"], errors='coerce')
sem_6_2015_df['ACN_TW'] = pd.to_numeric(sem_6_2015_df["ACN_TW"], errors='coerce')
sem_6_2015_df['ACN_P&O'] = pd.to_numeric(sem_6_2015_df["ACN_P&O"], errors='coerce')

sem_6_2015_df['SPCC'] = pd.to_numeric(sem_6_2015_df["SPCC"], errors='coerce')
sem_6_2015_df['SPCC_TW'] = pd.to_numeric(sem_6_2015_df["SPCC_TW"], errors='coerce')
sem_6_2015_df['SPCC_P&O'] = pd.to_numeric(sem_6_2015_df["SPCC_P&O"], errors='coerce')

sem_6_2015_df['OOSE'] = pd.to_numeric(sem_6_2015_df["OOSE"], errors='coerce')
sem_6_2015_df['OOSE_TW'] = pd.to_numeric(sem_6_2015_df["OOSE_TW"], errors='coerce')
sem_6_2015_df['OOSE_P&O'] = pd.to_numeric(sem_6_2015_df["OOSE_P&O"], errors='coerce')

sem_6_2015_df['AMP'] = pd.to_numeric(sem_6_2015_df["AMP"], errors='coerce')
sem_6_2015_df['AMP_TW'] = pd.to_numeric(sem_6_2015_df["AMP_TW"], errors='coerce')
sem_6_2015_df['AMP_P&O'] = pd.to_numeric(sem_6_2015_df["AMP_P&O"], errors='coerce')

sem_6_2015_df['DWM'] = pd.to_numeric(sem_6_2015_df["DWM"], errors='coerce')
sem_6_2015_df['DWM_TW'] = pd.to_numeric(sem_6_2015_df["DWM_TW"], errors='coerce')
sem_6_2015_df['DWM_P&O'] = pd.to_numeric(sem_6_2015_df["DWM_P&O"], errors='coerce')

sem_6_2015_df['SEMR'] = pd.to_numeric(sem_6_2015_df["SEMR"], errors='coerce')
sem_6_2015_df['SEMR_P&O']= pd.to_numeric(sem_6_2015_df["SEMR_P&O"], errors='coerce')

sem_6_2015_df['TOT VI'] = pd.to_numeric(sem_6_2015_df["TOT VI"], errors='coerce')
sem_6_2015_df['TOT V'] = pd.to_numeric(sem_6_2015_df["TOT V"], errors='coerce')
sem_6_2015_df['GNT'] = pd.to_numeric(sem_6_2015_df["GNT"], errors='coerce')
```
