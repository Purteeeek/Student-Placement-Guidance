```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
% matplotlib inline
```

```python
sem3_2018 = '../../Cleaned-Data/2018-Passouts/Sem-3_2018_Passout_cleaned.xlsx'
sem4_2018 = '../../Cleaned-Data/2018-Passouts/Sem-4_2018_Passout_cleaned.xlsx'
sem5_2018 = '../../Cleaned-Data/2018-Passouts/Sem-5_2018_Passout_cleaned.xlsx'
sem6_2018 = '../../Cleaned-Data/2018-Passouts/Sem-6_2018_Passout_cleaned.xlsx'
```
